# walker_run

```python
ENV_NAME = 'walker_run'
    
DICTS_ALL_STATS[ENV_NAME] = {}

datasets_dict, list_of_root_dirs_by_augment_degree = proc.fetch_data(
    EXP_TYPE, ENV_NAME, EXP_NAME, LIST_VALS, show=False)

column_name = 'episode_reward'
MIN_TOP_FIVE, MAX_TOP_FIVE, summary_statistics, sorted_items = proc.display_analysis(
    datasets_dict, list_of_root_dirs_by_augment_degree, ENV_NAME, column_name, show=False)

DICTS_ALL_STATS[ENV_NAME][column_name+'_sorted'] = sorted_items

print('TOP PERFORMANCE BETWEEN ', MIN_TOP_FIVE, ' AND ', MAX_TOP_FIVE)
proc.plot_performance_all(summary_statistics, ENV_NAME, column_name, XLABEL_STATS)


datasets_augstats_dict, list_of_root_dirs_by_augment_stats = proc.fetch_data_aug_stats(
    EXP_TYPE, ENV_NAME, EXP_NAME, LIST_VALS, show=False)

for stat_name in stats_column_names:
    sorted_mean_diststats = proc.plot_aug_stats(
        datasets_augstats_dict, 
        list_of_root_dirs_by_augment_stats, 
        stat_name, 
        MIN_TOP_FIVE, 
        MAX_TOP_FIVE, 
        XLABEL_STATS,
        show=True)
    DICTS_ALL_STATS[ENV_NAME][stat_name] = sorted_mean_diststats
```

```python
DICTS_AGG_ALL = {}

for k, v in DICTS_ALL_STATS.items():
    print('environment: ', k)
    DICTS_AGG_ALL[k] = {}
    aug_param_for_max_mean_reward = -1
    for ek, ev in DICTS_ALL_STATS[k].items():
        if ek == 'episode_reward_sorted':
            aug_param_for_max_mean_reward = DICTS_ALL_STATS[k][ek][0][0]
            print("HIGHEST MEAN SUM OBTAINED FOR AUGMENT PARAM: ", aug_param_for_max_mean_reward,
                 "(mean sum reward - ", DICTS_ALL_STATS[k][ek][0][1], ")")
            print("\n")
        else:
            tuple_stats_for_max_reward = [i for i in DICTS_ALL_STATS[k][ek] if i[0] == aug_param_for_max_mean_reward]
            max_value = max(i[1] for i in  DICTS_ALL_STATS[k][ek])
            min_value = min(i[1] for i in  DICTS_ALL_STATS[k][ek])
            print(tuple_stats_for_max_reward)
            DICTS_AGG_ALL[k][ek] = (tuple_stats_for_max_reward[0], min_value, max_value)
        
    print("===================================")
    
# print(DICTS_AGG_ALL)
env_width = max(len(env) for env in DICTS_AGG_ALL.keys()) + 2
stat_width = max(len(stat) for stat in stats_column_names) + 2
value_width = 25  # Adjust width for the tuple

header = f"{'Environment'.ljust(env_width)}{'StatDist'.ljust(stat_width)}{'DistForMaxReward'.ljust(value_width)}{'DistMin'.ljust(value_width)}{'DistMax'.ljust(value_width)}"
print('-' * len(header))
print('-' * len(header))
print(header)
print('-' * len(header))

    
for stats in stats_column_names:
    print("\n")
    mean_vals = []
    max_vals = []
    min_vals = []
    categories = []
    for envs in list(DICTS_AGG_ALL.keys()):
        value = DICTS_AGG_ALL[envs][stats]
        print(f"{envs.ljust(env_width)}{stats.ljust(stat_width)}{str(value[0][1]).ljust(value_width)}{str(value[1]).ljust(value_width)}{str(value[2]).ljust(value_width)}")
        mean_vals.append(value[0][1])
        max_vals.append(value[2])
        min_vals.append(value[1])
        categories.append(f"{envs} \n{stats} ({value[0][0]})")
    
    # Plot the values
    plt.figure(figsize=(12, 8))
    category_indices = np.arange(len(categories))
    plt.errorbar(
        category_indices, 
        mean_vals, 
        yerr=[np.array(mean_vals) - np.array(min_vals), 
        np.array(max_vals) - np.array(mean_vals)], 
        fmt='o', ecolor='black', 
        capsize=5, 
        label='Mean Distance with Min-Max Range')
    plt.xticks(category_indices, categories, rotation=45)
    plt.title(f"Distance Metrics for [{stats} metrics] giving Max Environment Performance, \nwith Min and Max Distance Values Obtained on Full Augmentation Range ({XLABEL_STATS})")
    plt.xlabel('Environment - Distance Metrics (Padding Value for Top Mean Sum of Episodic Rewards)')
    plt.ylabel('Statistical Distance Units')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
        
        
        
for stats in stats_column_names:
    print("\n")
    second_vals = []
    for envs in list(DICTS_AGG_ALL.keys()):
        value = DICTS_AGG_ALL[envs][stats]
        print(f"{envs.ljust(env_width)}{stats.ljust(stat_width)}{str(value[0][1]).ljust(value_width)}{str(value[1]).ljust(value_width)}{str(value[2]).ljust(value_width)}")
        second_vals.append(value[1])
    
    # Plot the values
    plt.figure(figsize=(10, 6))
    plt.scatter(second_vals, second_vals, color='blue', alpha=0.7)
    plt.title(f"Distribution of Distance Metrics Values for {stats} \nOver All {len(DICTS_AGG_ALL.keys())} Tested Environments")
    plt.xlabel('Statistical Distance Units')
    plt.ylabel('Statistical Distance Units')
    plt.grid(True)

    # Optionally, add a diagonal line to emphasize the identity line
    plt.plot([min(second_vals), max(second_vals)], [min(second_vals), max(second_vals)], "r--")

    plt.show()
```
