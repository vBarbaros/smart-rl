```python
import pandas as pd
import os
import glob

def find_files(root_dir, pattern):
    # Create a search pattern that includes the root directory and the subdirectory pattern
    search_pattern = os.path.join(root_dir, pattern)
    
    # List to store the file paths
    files_found = []
    
    # Use glob.glob to find all files that match the search pattern
    for file_path in glob.glob(search_pattern, recursive=True):
        if os.path.isfile(file_path):  # Ensure it's a file, not a directory
            files_found.append(file_path)
    
    return files_found



pendswing_pad_0_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pixels-True/seed_2/2024-05-06_14-12-56/eval.csv'
df_pswing_pad_0_seed_2_eval = pd.read_csv(pendswing_pad_0_seed_2_eval_path)
print(df_pswing_pad_0_seed_2_eval.head())


pendswing_pad_1_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pad_1_pixels-True/seed_2/2024-05-06_14-12-55/eval.csv'
df_pswing_pad_1_seed_2_eval = pd.read_csv(pendswing_pad_1_seed_2_eval_path)
print(df_pswing_pad_1_seed_2_eval.head())


pendswing_pad_2_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pad_2_pixels-True/seed_2/2024-05-06_14-12-58/eval.csv'
df_pswing_pad_2_seed_2_eval = pd.read_csv(pendswing_pad_2_seed_2_eval_path)
print(df_pswing_pad_2_seed_2_eval.head())


pendswing_pad_3_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pad_3_pixels-True/seed_2/2024-05-06_14-12-57/eval.csv'
df_pswing_pad_3_seed_2_eval = pd.read_csv(pendswing_pad_3_seed_2_eval_path)
print(df_pswing_pad_3_seed_2_eval.head())


pendswing_pad_4_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pad_4_pixels-True/seed_2/2024-05-06_14-12-55/eval.csv'
df_pswing_pad_4_seed_2_eval = pd.read_csv(pendswing_pad_4_seed_2_eval_path)
print(df_pswing_pad_4_seed_2_eval.head())


pendswing_pad_5_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pad_5_pixels-True/seed_2/2024-05-06_14-12-55/eval.csv'
df_pswing_pad_5_seed_2_eval = pd.read_csv(pendswing_pad_5_seed_2_eval_path)
print(df_pswing_pad_5_seed_2_eval.head())


pendswing_pad_6_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pad_6_pixels-True/seed_2/2024-05-06_14-12-55/eval.csv'
df_pswing_pad_6_seed_2_eval = pd.read_csv(pendswing_pad_6_seed_2_eval_path)
print(df_pswing_pad_6_seed_2_eval.head())


pendswing_pad_7_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pad_7_pixels-True/seed_2/2024-05-06_14-12-55/eval.csv'
df_pswing_pad_7_seed_2_eval = pd.read_csv(pendswing_pad_7_seed_2_eval_path)
print(df_pswing_pad_7_seed_2_eval.head())


pendswing_pad_8_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pad_8_pixels-True/seed_2/2024-05-06_14-12-55/eval.csv'
df_pswing_pad_8_seed_2_eval = pd.read_csv(pendswing_pad_8_seed_2_eval_path)
print(df_pswing_pad_8_seed_2_eval.head())


pendswing_pad_9_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pad_9_pixels-True/seed_2/2024-05-06_14-12-54/eval.csv'
df_pswing_pad_9_seed_2_eval = pd.read_csv(pendswing_pad_9_seed_2_eval_path)
print(df_pswing_pad_9_seed_2_eval.head())


pendswing_pad_10_seed_2_eval_path = '../exp-pad/exp/pendulum_swingup/drqv2_pad_10_pixels-True/seed_2/2024-05-06_14-12-55/eval.csv'
df_pswing_pad_10_seed_2_eval = pd.read_csv(pendswing_pad_10_seed_2_eval_path)
print(df_pswing_pad_10_seed_2_eval.head())


```
